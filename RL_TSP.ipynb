```python
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
from numpy import random
import pandas as pd
from scipy.spatial import distance_matrix
```

```python
n_dest = 20
```

```python
x = np.random.uniform(0, 100, n_dest)
y = np.random.uniform(0, 100, n_dest)
cities = pd.DataFrame()
cities['x']=x
cities['y']=y
cities
```

```python
cities.plot(kind='scatter', x='x',y='y')
plt.grid()
```

```python
dist = pd.DataFrame(distance_matrix(cities.values, cities.values))
neg_dist = -1*dist
```

```python
dist
```

## Q-Learning

```python
def update_q(q, state, action, alpha=0.01, gamma=0.4):
    reward = neg_dist[state][action]
    q[state,action] = q[state,action]+alpha*(reward+gamma*q[action,:].max()-q[state,action]) #greedy policy
    return q
```

```python
q = np.zeros([n_dest,n_dest])
n_train = 10000
q_learning_result=[]
for i in range(n_train):
    epsilon = 1-i/n_train
    traj = [0]
    state = 0
    distance_travel = 0
    possible_actions = [dest for dest in range(n_dest) if dest not in traj]
    while possible_actions:
        if random.random() < epsilon:
            action = random.choice(possible_actions)
        else:       
            best_action_index = q[state, possible_actions].argmax()
            action = possible_actions[best_action_index]
        distance_travel += dist[state][action]
        possible_actions.remove(action)
        traj.append(action)
        q = update_q(q, state, action)
        state = action
        if len(possible_actions) == 0:  
            state=action
            action=0
            distance_travel += dist[state][action]
            traj.append(action)
    q_learning_result.append([i,distance_travel])
```

```python
result1=pd.DataFrame(q_learning_result, columns=['number','Total Distance'])
```

```python
ax = result1.plot(x='number',y='Total Distance',figsize=(20,8))
ax.set_xlabel('Number of Iteration', fontsize=15)
ax.set_ylabel('Total Travel Distance', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize =15)
plt.savefig('Q_inverse.png')
```

```python
print('Best trajectory found:')
print(' -> '.join([str(b) for b in traj]))
print(f'Inverse Distance Travelled: {distance_travel}')
```

## SARSA

```python
def update_q(q, state, action, next_action, alpha=0.01, gamma=0.4):
    reward = neg_dist[state][action]
    q[state,action] = q[state,action]+alpha*(reward+gamma*q[action,next_action]-q[state,action])
    return q
```

```python
q = np.zeros([n_dest,n_dest])
n_train = 10000
SARSA_result=[]
for i in range(n_train):
    epsilon = 1-i/n_train
    traj = [0]
    state = 0
    distance_travel = 0
    possible_actions = [dest for dest in range(n_dest) if dest not in traj]
    while possible_actions:
        if random.random() < epsilon:
            action = random.choice(possible_actions)
        else:       
            best_action_index = q[state, possible_actions].argmax()
            action = possible_actions[best_action_index]
        distance_travel += dist[state][action]
        possible_actions.remove(action)
        traj.append(action)
        if len(possible_actions) != 0:
            if random.random() < epsilon:
                next_action = random.choice(possible_actions)
            else:       
                best_action_index = q[action, possible_actions].argmax()
                next_action = possible_actions[best_action_index]        
            q = update_q(q, state, action, next_action)
            state = action
        else:
            state=action
            action=0
            distance_travel += dist[state][action]
            traj.append(0)
    SARSA_result.append([i,distance_travel])
```

```python
result2=pd.DataFrame(SARSA_result, columns=['number','Total Distance'])
```

```python
ax = result2.plot(x='number',y='Total Distance',figsize=(20,8))
ax.set_xlabel('Number of Iteration', fontsize=15)
ax.set_ylabel('Total Travel Distance', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize =15)
plt.savefig('SARSA_inverse.png')
```

```python
print('Best trajectory found:')
print(' -> '.join([str(b) for b in traj]))
print(f'Inverse Distance Travelled: {distance_travel}')
```

```python
ax = result2.plot(x='number',y='Total Distance',figsize=(20,8))
ax.set_xlabel('Number of Iteration', fontsize=15)
ax.set_ylabel('Total Travel Distance', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize =15)
plt.savefig('SARSA_negative.png')
```

## MC prediction

```python
def epi_one(q, epsilon):
    traj = [0]
    state=0
    actions=[]
    rewards=[]
    states=[]
    possible_actions = [dest for dest in range(n_dest) if dest not in traj]
    while possible_actions:
        if random.random() < epsilon:
            action = random.choice(possible_actions)
        else:       
            best_action_index = q[state, possible_actions].argmax()
            action = possible_actions[best_action_index]
        traj.append(action)
        reward = 1/dist[state][action]
        rewards.append(reward)
        states.append(state)
        actions.append(action)
        state=action
        possible_actions.remove(action)
        if len(possible_actions) == 0:
            reward = 1/dist[action][0]
            rewards.append(reward)
            states.append(state)
            traj.append(0)
            actions.append(0)
    return rewards, states, actions, traj
```

```python
def calculate_G(rewards, states, gamma=0.4):
    rewards.reverse()
    states.reverse()
    G=[]
    for idx, state in enumerate(states):
        if idx==0:
            G.append(rewards[idx])
        else:
            G.append(rewards[idx] + gamma*G[idx-1])  
    rewards.reverse()
    states.reverse()  
    G.reverse()
    return G
```

```python
def update_q(q, G, states, actions, epi):
    for i in range(len(states)):
        state = states[i]
        action = actions[i]
        q[state,action] = q[state,action]+1/(epi+1)*(G[i] - q[state,action])
    return q
```

```python
episodes = 10000
q = np.zeros([n_dest,n_dest])
MC_result=[]
for epi in range(episodes):
    epsilon = 1-epi/episodes
    rewards, states, actions, traj = epi_one(q, epsilon)
    G = calculate_G(rewards, states)
    q = update_q(q, G, states, actions, epi)
    distance_travel=0
    for i in range(n_dest):
        distance_travel += dist[traj[i]][traj[i+1]]
    MC_result.append([epi, distance_travel])
```

```python
result3=pd.DataFrame(MC_result, columns=['number','Total Distance'])
```

```python
ax = result3.plot(x='number',y='Total Distance',figsize=(20,8))
ax.set_xlabel('Number of Episodes', fontsize=15)
ax.set_ylabel('Total Travel Distance', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize =15)
plt.savefig('MC_inverse.png')
```

```python
print('Best trajectory found:')
print(' -> '.join([str(b) for b in traj]))
print(f'Inverse Distance Travelled: {distance_travel}')
```
